# Split Dataset

```python
from sklearn.model_selection import train_test_split

# Split the data into features (X) and target variable (y)
X = traffic_data.drop("Traffic Situation", axis=1) 
y = traffic_data["Traffic Situation"]

# Perform a random train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1711)

print(f'There are {traffic_data.shape[0]} rows and {traffic_data.shape[1]} columns')

# Print the shape of the training and testing sets
print("Shape of Training Set: X_train:{0} and y_train:{1}".format(X_train.shape, y_train.shape))
print("Shape of Testing Set: X_test:{0} and y_test:{1}".format(X_test.shape, y_test.shape))
```

# Data Preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import RandomizedSearchCV


# Define numerical and categorical features
numerical_features = ['Date', 'CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total', 'Hour', 'Minute']
categorical_features = ['Day of the week']


# Define preprocessing steps for numerical and categorical features
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehotencoder', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps for numerical and categorical features
preprocessor = ColumnTransformer(transformers=[
    ('numerical', numerical_transformer, numerical_features),
    ('categorical', categorical_transformer, categorical_features)
])
```

#  Build Model with Preprocessing Pipeline

## 1. Hyperparameter Tuning

```python
# Define the model
model = RandomForestClassifier(random_state=1711)

# Create a pipeline with preprocessing and the model
pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('randomforestclassifier', model)])

# Define the hyperparameter grid for RandomizedSearchCV
param_grid = {
    'randomforestclassifier__n_estimators': list(range(60, 105)),
    'randomforestclassifier__criterion': ['gini', 'entropy'],
    'randomforestclassifier__max_depth': list(range(1, 10)),
    'randomforestclassifier__max_features': ['sqrt', 'log2'],
    'randomforestclassifier__min_samples_split': list(range(2, 10)),
    'randomforestclassifier__min_samples_leaf': list(range(1, 5)),
    'randomforestclassifier__bootstrap': [True, False]
}

# Define cross-validation strategy
kfold = KFold(n_splits=5, shuffle=True, random_state=1711)

# Perform RandomizedSearchCV
rd_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_grid,
    scoring='accuracy',  # or any other appropriate metric for classification - f1_macro', 'precision_macro', or 'recall_macro'
    cv=kfold,
    n_iter=100,  # Adjust the number of iterations as needed
    random_state=1711
)

# Fit the RandomizedSearchCV on the training data
rd_search.fit(X_train, y_train)
```

```python
print("Validation Accuracy: %.2f %%" % (rd_search.best_score_*100))
```

```python
print("Accuracy on train+val data: %.2f %%" % (rd_search.score(X_train, y_train)*100))
```

## 2. Best Estimator

```python
# Access the best estimator from the randomized search
best_estimator = rd_search.best_estimator_
best_estimator
```

```python
# Use the best estimator to make predictions on the test set
y_pred = best_estimator.predict(X_test)
y_pred
# y_test.head(10)
```
